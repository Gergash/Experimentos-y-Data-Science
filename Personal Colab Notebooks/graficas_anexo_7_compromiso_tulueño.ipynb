```python
import matplotlib.pyplot as plt

# Datos para las gráficas
# 1. Distribución de compromiso
commitment_labels = ["Muy comprometido", "Comprometido", "Para nada comprometido"]
commitment_values = [29, 10, 17]

# 2. Preferencias por áreas sociales
areas_labels = ["Civismo", "Educación", "Arte", "Deporte", "Seguridad"]
areas_values = [16, 14, 11, 8, 7]

# 3. Importancia de la colaboración
collaboration_labels = ["Muy importante", "Importante", "Poco importante"]
collaboration_values = [20, 17, 19]

# 4. Comportamiento en espacios públicos
behavior_labels = ["Muy respetuoso", "Respetuoso", "Indiferente"]
behavior_values = [22, 18, 16]

# Gráfica 1: Distribución de compromiso
plt.figure(figsize=(8, 6))
plt.bar(commitment_labels, commitment_values, color="steelblue", alpha=0.7)
plt.title("Distribución del Compromiso", fontsize=16)
plt.ylabel("Número de Respuestas", fontsize=12)
plt.xlabel("Nivel de Compromiso", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()

```

```python

# Gráfica 2: Preferencias por áreas sociales
plt.figure(figsize=(8, 6))
plt.bar(areas_labels, areas_values, color="teal", alpha=0.7)
plt.title("Preferencias por Áreas Sociales", fontsize=16)
plt.ylabel("Número de Respuestas", fontsize=12)
plt.xlabel("Áreas Sociales", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python

# Gráfica 3: Importancia de la colaboración
plt.figure(figsize=(8, 6))
plt.pie(collaboration_values, labels=collaboration_labels, autopct='%1.1f%%', startangle=90, colors=["gold", "skyblue", "lightcoral"])
plt.title("Importancia de la Colaboración", fontsize=16)
plt.show()
```

```python

# Gráfica 4: Comportamiento en espacios públicos
plt.figure(figsize=(8, 6))
plt.pie(behavior_values, labels=behavior_labels, autopct='%1.1f%%', startangle=90, colors=["limegreen", "cornflowerblue", "tomato"])
plt.title("Comportamiento en Espacios Públicos", fontsize=16)
plt.show()
```
