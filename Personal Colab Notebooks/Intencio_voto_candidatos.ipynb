```python
# Agregar todas las filas de las distintas columnas 'Modelado' en los distintos datasets a un dataframe aparte
intencion_voto_fechas = pd.concat([data_Respuestas["Registro"], data_Conocimiento["Registro"], data_Repes_final_elecciones["Registro"]], axis=0)
# Renombrar la columna a 'Modelado_Predictivo'
intencion_voto_fechas
```

```python
intencion_voto_candidatos = pd.concat([data_Respuestas["2. Si la elección fuera hoy, ¿por cuál de los siguientes candidatos oficiales votaría?"], data_Conocimiento["7. En su opinión, ¿qué candidato/a tiene las mejores propuestas o soluciones para abordar los problemas actuales del país/estado/municipio?"], data_Repes_final_elecciones['1. ¿Cuál de los siguientes candidatos cree usted que tiene las capacidades para<b> cumplir</b> con su programa de gobierno?']], axis=0)
intencion_voto_candidatos
```

```python
# Supongamos que tienes dos Series 'serie_registro' y 'serie_candidato'

# Crear un nuevo DataFrame combinando las series
data = {'Registro': intencion_voto_fechas, 'Candidato': intencion_voto_candidatos}
intencion_voto = pd.DataFrame(data)

# Si deseas guardar el DataFrame resultante en un archivo CSV, por ejemplo:
# intencion_voto.to_csv("ruta_del_archivo_intencion_voto.csv", index=False)
```

#3-) Limpieza de los registros

```python
# Buscar todas las filas con el nombre "Jose Alexander Gallego Chavez" o "Jorge Alexander Gallego Chávez"
mask = (intencion_voto["Candidato"] == "Jose Alexander Gallego Chávez") | (intencion_voto["Candidato"] == "Jorge Alexander Gallego Chávez")

# Cambiar el nombre por "Jorge Alexander Gallego Chavez"
intencion_voto.loc[mask, "Candidato"] = "Jorge Gallego Chavez"

# Buscar todas las filas con el nombre "Gustavo Adolfo Vélez Román"
mask = intencion_voto["Candidato"] == "Gustavo Adolfo Vélez Román"

# Cambiar el nombre por "Gustavo Velez Roman"
intencion_voto.loc[mask, "Candidato"] = "Gustavo Velez Roman"

# Buscar todas las filas con el nombre "Evert Antonio Villegas Morante" o "Ever Antonio Villegas Morante"
mask = (intencion_voto["Candidato"] == "Evert Antonio Villegas Morante") | (intencion_voto["Candidato"] == "Ever Antonio Villegas Morante")

# Cambiar el nombre por "Evert Villegas Morante"
intencion_voto.loc[mask, "Candidato"] = "Evert Villegas Morante"

# Buscar todas las filas con el nombre "Marco Alejandro Arbeláez Vargas" o "Marco Alejando Árbelaez Vargas"
mask = (intencion_voto["Candidato"] == "Marco Alejandro Arbeláez Vargas") | (intencion_voto["Candidato"] == "Marco Alejando Árbelaez Vargas")

# Cambiar el nombre por "Marco Arbelaez Vargas"
intencion_voto.loc[mask, "Candidato"] = "Marco Arbelaez Vargas"
```
