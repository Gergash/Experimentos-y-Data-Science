# Regresion Lineal Multiple

```python
data_regresion_lineal_multiple= data[['NA_Sales', 'JP_Sales', 'Global_Sales']]
data_regresion_lineal_multiple.head()
```

```python
import pandas as pd                                # Se trabaja todo lo que es la estructura de los datos
import statsmodels.formula.api as smf              # Permite ajustar modelos estad√≠sticos utilizando f√≥rmulas de estilo R
import numpy as np                                 # representa matrices multidimensionales, equivalentes a las matrices del R
from matplotlib import pyplot                      # Permite la generaci√≥n de gr√°ficos
from mpl_toolkits.mplot3d import Axes3D            # Permite agregar eje tridimensionales
from sklearn.linear_model import LinearRegression  # sklearn es una libreria que cuenta con algoritmos de clasificaci√≥n, regresi√≥n, clustering y reducci√≥n de dimensionalidad
from sklearn import metrics                        # sklearn es una libreria que cuenta con algoritmos de clasificaci√≥n, regresi√≥n, clustering y reducci√≥n de dimensionalidad
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Suponiendo que 'data' es tu DataFrame que contiene las columnas 'NA_Sales', 'JP_Sales' y 'Global_Sales'

# Convertir las columnas de tipo float a int si es necesario
data['NA_Sales'] = data['NA_Sales'].astype(int)
data['JP_Sales'] = data['JP_Sales'].astype(int)
data['Global_Sales'] = data['Global_Sales'].astype(int)

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Extraer datos de las columnas
x1 = data['NA_Sales']      # Ventas en Norte Am√©rica (eje X)
x2 = data['JP_Sales']      # Ventas en Jap√≥n (eje Y)
y = data['Global_Sales']   # Ventas Globales (eje Z)

# Graficar el diagrama de dispersi√≥n tridimensional
ax.scatter(x1, x2, y, marker='*', c='r')

# Etiquetas de los ejes
ax.set_xlabel('Ventas en Norte Am√©rica')
ax.set_ylabel('Ventas en Jap√≥n')
ax.set_zlabel('Ventas Globales')

# Mostrar la gr√°fica
plt.show()
```

```python
data_regresion_lineal_multiple.corr()   # Podemos observar un alto indice de correlacion entre cada variable
```

```python
reg = smf.ols('NA_Sales ~ JP_Sales + Global_Sales',data=data_regresion_lineal_multiple)   # Se establece la regrecion con el Ordinary Least Square dictando que el consumo
res = reg.fit()                                                                           # va a ser aproximandamente igual a la suma del cilindraje y potencia.
```

```python
print(res.summary())    # Imprimimos los resultados
print(res.rsquared)
print(res.params)
```

```python
lm = LinearRegression()
X = data_regresion_lineal_multiple[['NA_Sales','JP_Sales']]
lm.fit(X,y)
prediccion = lm.predict(X)
a = lm.intercept_
b = lm.coef_
```

```python
b
```

```python
print('Prediccion')
print('*'*100)
print(prediccion)     # Valores segun la prediccion
print('*'*100)
print(y)              # Valores actuales del consumo
print(lm.score(X,y))  # Retornamos el coeficiente de determinacion en la prediccion
```

```python
resultados ={'Real':data_regresion_lineal_multiple['Global_Sales'],'Prediccion':prediccion}
result = pd.DataFrame(data=resultados)
print(result)
```

```python
ECM = metrics.mean_squared_error(data_regresion_lineal_multiple['Global_Sales'],prediccion)
print(ECM)         # Error estandar
r_cuadrado = metrics.r2_score(data_regresion_lineal_multiple['Global_Sales'],prediccion)
print(r_cuadrado)  # R cuandrado o coeficiente de determinacion
```

```python
# Crear la figura y el eje 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Extraer datos de las columnas
x1 = data['NA_Sales']      # Ventas en Norte Am√©rica (eje X)
x2 = data['JP_Sales']      # Ventas en Jap√≥n (eje Y)
y = data['Global_Sales']   # Ventas Globales (eje Z)

# Graficar el diagrama de dispersi√≥n tridimensional
ax.scatter(x1, x2, y, marker='*', c='r')
# Datos segun la prediccion
ax.scatter(x1, x2, prediccion, marker='+', c='b')
# Etiquetas de los ejes
ax.set_xlabel('Ventas en Norte Am√©rica')
ax.set_ylabel('Ventas en Jap√≥n')
ax.set_zlabel('Ventas Globales')

# Mostrar la gr√°fica
plt.show()
```
